## ETAPA 3: Entrenamiento de Modelos de Clasificaci√≥n y Evaluaci√≥n de la Calidad<a id=train></a>

### √Årboles de Decisi√≥n - Hiperpar√°metros (`Valores por Defecto`)<a id='dt1'></a>

```python
# Para esta primera parte de Entrenamiento y Evaluaci√≥n de los modelos utilizar√© el algoritmo "DecisionTreeClassifier"
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# Defino un modelo del tipo de DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=12345)

# Entrenamos el modelo DecisionTreeClassifier con el conjunto de entrenamiento
dt_model.fit(X_train, y_train)

# Seguidamente realizo la evaluaci√≥n del modelo, utilizando el Conjunto de Validaci√≥n
y_val_pred_dt1 = dt_model.predict(X_valid)

accuracy1 = accuracy_score(y_valid, y_val_pred_dt1)

print(f"La precisi√≥n obtenida con los valores de los hiperpar√°metros por defecto es: {accuracy1:.4f}")
```

<span style='color:blue'>
    
### √Årboles de Decisi√≥n - *Hiperpar√°metro (`max_depth`)*<a id='dt2'></a>
    
</span>


```python
# Ahora buscar√© optimizar la precisi√≥n del modelo mediante el ajuste del hiperpar√°metro "max_depth"
best_accuracy = 0
best_depth = 0

for depth in range(1, 31):
    dt_model = DecisionTreeClassifier(max_depth=depth, random_state=12345)
    dt_model.fit(X_train, y_train)
    y_val_pred_dt2 = dt_model.predict(X_valid)
    accuracy2 = accuracy_score(y_valid, y_val_pred_dt2)
    
    if accuracy2 > best_accuracy:
        best_accuracy = accuracy2
        best_depth = depth
        
print(f"Mejor ajuste del hiperpar√°metro MAX_DEPTH={best_depth}   -   PRECISION={best_accuracy:.4f}")
```

<span style='color:blue'>
    
### √Årboles de Decisi√≥n - *Hiperpar√°metro (`min_samples_split`)*<a id='dt3'></a>
    
</span>


```python
# Ajustaremos el hiperpar√°metro 'min_samples_split' y observaremos c√≥mo afecta la precisi√≥n
best_accuracy_split = 0
best_split = 0

# Probar diferentes valores de min_samples_split
for split in range(2, 31):
    dt_model = DecisionTreeClassifier(max_depth=7, min_samples_split=split, random_state=12345)
    dt_model.fit(X_train, y_train)
    y_val_pred_dt3 = dt_model.predict(X_valid)
    accuracy3 = accuracy_score(y_valid, y_val_pred_dt3)
    
    if accuracy3 > best_accuracy_split:
        best_accuracy_split = accuracy3
        best_split = split

print(f"Mejor ajuste del hiperpar√°metro MIN_SAMPLES_SPLIT={best_split}   -   PRECISION={best_accuracy_split:.4f}")
```

### Generalidades de las Pruebas de Cordura<a id='cordura1'></a>
    
</span>

Las pruebas de cordura (*sanity checks*) son una t√©cnica utilizada en machine learning para **asegurarse de que el modelo se comporte de manera razonable antes de realizar evaluaciones m√°s profundas o implementar el modelo en un entorno de producci√≥n**. Estas pruebas no buscan medir el rendimiento final del modelo, sino asegurarse de que el modelo no est√© haciendo suposiciones err√≥neas o tomando decisiones triviales.

---

<span style='color:brown'>

**¬øPara qu√© sirven las pruebas de cordura?**

</span>

 1. **Verificar el comportamiento b√°sico del modelo**: Las pruebas de cordura ayudan a garantizar que el modelo no est√© tomando decisiones sin sentido o basadas en artefactos triviales en los datos.

 2. **Evitar errores de implementaci√≥n**: A veces, un modelo puede parecer estar funcionando bien, pero una prueba de cordura puede revelar errores de codificaci√≥n o malentendidos en los datos.

 3. **Validar datos y etiquetado**: Tambi√©n son √∫tiles para detectar problemas en los datos, como etiquetados incorrectos o una segmentaci√≥n incorrecta entre las caracter√≠sticas y las etiquetas.

 4. **Prevenir resultados enga√±osos**: Las pruebas de cordura aseguran que un modelo no est√° obteniendo resultados buenos por motivos incorrectos, como una fuga de datos (*cuando el modelo tiene acceso a informaci√≥n que no deber√≠a durante el entrenamiento*).
 
 
---

<span style='color:brown'>

**¬øC√≥mo se realizan las pruebas de cordura?**
    
</span>

Algunos enfoques comunes para realizar pruebas de cordura incluyen:

 1. **Modelo trivial (predicci√≥n constante):**

     * Se entrena un modelo trivial que hace predicciones constantes para todas las observaciones (*por ejemplo, siempre predice la clase mayoritaria o una clase aleatoria*). Si tu modelo tiene una precisi√≥n similar al de este modelo trivial, entonces algo est√° mal.
     
     * **Ejemplo**: Predecir siempre la clase "0" (por ejemplo, Smart) y comparar la precisi√≥n de este modelo con la precisi√≥n del modelo entrenado.


 2. **Entrenamiento con datos aleatorizados:**

     * Se reordenan aleatoriamente las etiquetas (**`y`**) y se entrena el modelo. Si el modelo sigue obteniendo una precisi√≥n relativamente alta, podr√≠a indicar una fuga de datos o un error.
     
     * Al entrenar con etiquetas aleatorias, el modelo no deber√≠a poder aprender ninguna relaci√≥n significativa y deber√≠a obtener una precisi√≥n cercana al azar.


 3. **Comprobaci√≥n de la consistencia de las caracter√≠sticas:**

     * Se puede entrenar el modelo con solo una caracter√≠stica aleatoria o irrelevante, y si la precisi√≥n es razonable, podr√≠a haber un problema con el conjunto de datos o el etiquetado.
     

 4. **Predicci√≥n con datos vac√≠os o caracter√≠sticas sin sentido:**
 
     * Se puede entrenar el modelo con datos vac√≠os o caracter√≠sticas irrelevantes (*como una constante*) para ver si el modelo reacciona de manera l√≥gica.
     
     
 5. **Cross-validation simple:**

     * Realizar una validaci√≥n cruzada sencilla para comprobar que el modelo no est√° sobreajustando o subajustando en diferentes particiones de los datos.

 2. **Entrenar el modelo con etiquetas aleatorias**: Entrenamos el modelo con las etiquetas desordenadas.

### Validaci√≥n Cruzada<a id='cordura6'></a>
    
</span>



---

<div style='background-color:lightyellow; padding:15px;'>

La validaci√≥n cruzada es una t√©cnica utilizada para evaluar el rendimiento de un modelo de manera m√°s robusta al entrenarlo y evaluarlo en diferentes particiones del conjunto de datos. **Es una prueba de cordura √∫til para verificar que el modelo no est√° sobreajustado o subajustado a una √∫nica partici√≥n de los datos**.

Para su implementaci√≥n los pasos a seguir son los que se detallan a continuaci√≥n:

 1. **Dividir los datos**: En la validaci√≥n cruzada, los datos se dividen en varios subconjuntos. **El modelo se entrena en un subconjunto y se eval√∫a en otro**. Este proceso se repite hasta que todos los subconjuntos hayan sido utilizados como conjunto de validaci√≥n al menos una vez.
    
    
 2. **Entrenar y evaluar el modelo**: Entrenamos el modelo de `RandomForestClassifier` en cada subconjunto y calculamos la precisi√≥n media a trav√©s de todos los subconjuntos.
    
    
 3. **Interpretaci√≥n**:
    
    - **Alta consistencia entre subconjuntos**: Si las puntuaciones de precisi√≥n son consistentes entre los diferentes subconjuntos, esto indica que el modelo generaliza bien.
    
    - **Variabilidad alta entre subconjuntos**: Si hay una alta variabilidad en la precisi√≥n entre diferentes subconjuntos, el modelo podr√≠a estar sobreajustado o subajustado.
    

**`Consideraciones para su Implementaci√≥n`**:
    
Vamos a implementar la validaci√≥n cruzada utilizando 5 "folds". Esto significa que los datos se dividir√°n en 5 subconjuntos, y en cada iteraci√≥n, 4 subconjuntos se usar√°n para entrenar y el subconjunto restante para validar. Al final, tomaremos la media de las precisiones obtenidas.
    
</div>

---

```python
from sklearn.model_selection import cross_val_score

# Entrenar el modelo de RandomForestClassifier utilizando validaci√≥n cruzada (5 folds)
rf_model_cv = RandomForestClassifier(n_estimators=200, max_depth=10, min_samples_split=9, random_state=12345)

# Aplicar validaci√≥n cruzada
cv_scores = cross_val_score(rf_model_cv, X_train, y_train, cv=5, scoring='accuracy')

# Calcular la precisi√≥n promedio y la desviaci√≥n est√°ndar
cv_mean_accuracy = np.mean(cv_scores)
cv_std_deviation = np.std(cv_scores)

print(f"Precisi√≥n Promedio: {cv_mean_accuracy:.4f}   -   Desviaci√≥n Est√°ndar: {cv_std_deviation:.4f}")
```

El modelo de `RandomForestClassifier` con validaci√≥n cruzada de 5 "folds" alcanz√≥ una **precisi√≥n promedio de 0.8237** y  una **desviaci√≥n est√°ndar de 0.0084**.

El valor de la `Precisi√≥n Promedio` indica que el modelo tiene un rendimiento consistente, con un valor cercano a lo obtenido en las pruebas anteriores.
